**Texto selecionado:** The Loves of the Lady Arabella, de Molly Elliot Seawell (Versão em língua inglesa)

O livro foi selecionado devido a sua fácil disponibilidade (livro em domínio público e formato txt disponível na web) e língua inglesa, que tornou mais simples a observação da frequência das palavras com base na teoria original apresentada.

Devido a extensão do texto, a apresentação gráfica foi propositalmente limitada às 100 maiores ocorrências.

```python
import urllib.request

texto = []

for line in urllib.request.urlopen('https://www.gutenberg.org/cache/epub/68033/pg68033.txt'):
  texto.append(line.decode('utf-8'))

for i in range(len(texto)):
  texto[i] = texto[i].lower()
  texto[i] = texto[i].replace('\n', '')
  texto[i] = texto[i].replace('.', '')
  texto[i] = texto[i].replace(',', '')
  texto[i] = texto[i].replace('(', '')
  texto[i] = texto[i].replace(')', '')
  texto[i] = texto[i].replace('?', '')
  texto[i] = texto[i].replace('\'', '')
```

```python
mydict = {}

for line in texto:
  line = line.lower()
  words = line.split()

  for word in words:
    if word not in mydict.keys():
      mydict[word] = 1    
    else:
      mydict[word] = mydict[word] + 1

print(mydict)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

df = pd.DataFrame(mydict.items(), columns=['word', 'count']).sort_values('count', ascending=False)
df = df[df['count'] > 4]
df = df.iloc[:100]

plt.figure(figsize=(24,10))
mpl.style.use(['seaborn'])
sns.barplot(x=df.word, y=df['count'])
plt.xticks(rotation=90)

plt.show()
```

**Conclusão:** O gráfico demonstra de forma inegável: a teoria permanece totalmente aplicável, demonstrando uma frequência muito maior de termos que não contribuem para a explicação do texto e sim apenas para possibilitar a leitura, como "the", "and", "to", "of e "i", que acredito serem os termos mais utilizados na imensa maioria dos textos em língua inglesa (no caso do último, em textos em primeira pessoa).
