#### Item d - Visualização em análise exploratória (Disponibilidade de filmes por país)

```python
x = titles_df.groupby(['country']).count()

x.head()

x.sort_values('show_id', ascending=False, inplace=True)
x.drop(x.iloc[:, 1:], inplace=True, axis=1)

x['show_id'].head(10).plot(kind='bar')
plt.title('Netflix - Disponibilidade de filmes por país')
plt.xlabel('País')
plt.ylabel('Quantidade de filmes')
```

### Parte 4 - Séries temporais

#### Quantidade de títulos por data de lançamento

```python
x = titles_df.groupby(['release_year']).count()

x.drop(x.iloc[:, 1:], inplace=True, axis=1)

plt.plot(x)
plt.title('Netflix - Disponibilidade de filmes por ano de lançamento')
plt.xlabel('Ano de lançamento')
plt.ylabel('Quantidade de filmes')

plt.show()
```
